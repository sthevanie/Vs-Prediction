```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
## Load Data ##
data = np.loadtxt("SANTANG-1 - Copy.las", skiprows=7198)
data[data==-999.2500]=np.nan
column = ['DEPTH','CALI','RHOB','DT','DTS','GR','LLD','LLS','MSFL','NPHI','PEF','SP']
data = pd.DataFrame(data, columns=column)
data = data.drop([7420],axis=0)
Depth = data[['DEPTH']].values

## Define Data ##
colm = ['RHOB','DT','GR','LLD','LLS','MSFL','NPHI','PEF','SP']
x = data[colm]
y = data[['DTS']].values
```

```python
## Split Data ##
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=0)

## Standardization ##
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
## Build Network ##
model = Sequential()

model.add(Dense(9,activation='relu'))

model.add(Dense(6,activation='relu'))
model.add(Dense(3,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
## Training ##
#early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=2, patience=10)
model.fit(x_train, y_train, validation_data=(x_test,y_test), batch_size=120,epochs=500, verbose = 2) #update weighting
# model.fit(x_train,y_train, validation_data=(x_test,y_test), batch_size=120, epochs=500, callbacks=[early_stop], verbose=2) #update weighting

# plt.plot(history.history['loss']) 
# plt.plot(history.history['val_loss']) 
# plt.title('Model loss') 
# plt.ylabel('Loss') 
# plt.xlabel('Epoch') 
# plt.legend(['Train', 'Test'], loc='upper left') 
# plt.show()
```

```python
losses = pd.DataFrame(model.history.history)
print(losses)
# print('min loss: ',min(losses['val_loss']))
losses.plot()
# plt.ylim(40,100)
plt.xlabel('Epoch')
plt.ylabel('Mean Square Error')
plt.title('Model Loss')
```

```python
from sklearn import metrics

Vs_train = model.predict(x_train)
Vs_test = model.predict(x_test)

tmae = metrics.mean_absolute_error(y_train,Vs_train)
trmse = np.sqrt(metrics.mean_squared_error(y_train,Vs_train))
mae = metrics.mean_absolute_error(y_test,Vs_test)
rmse = np.sqrt(metrics.mean_squared_error(y_test,Vs_test))
print('Validation Correlation')
print('#Train Evaluation#')
print('Mean Absolute Error: ',tmae)
print('Root Mean Square Error: ', trmse)
print()
print('#Test Evaluation#')
print('Mean Absolute Error: ',mae)
print('Root Mean Square Error: ', rmse)

plt.figure()
plt.scatter(y_test,Vs_test,color='r',label='data')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'k-',lw=2)
plt.title('Crossplot (Correlation Input)')
plt.xlabel('Vs Measured (us/ft)')
plt.ylabel('Vs Calculated (us/ft)')

plt.show()
```
